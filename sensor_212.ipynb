```python
import pandas as pd
history = pd.read_csv("data/full_sensor_data.csv", parse_dates=True, index_col="UTC time")
history.columns = pd.MultiIndex.from_tuples([(c.split('_')[0], c.split('_')[1]) for c in history.columns])
history.head()
```

```python
hist_212 = history['212']
hist_212.head()
```

```python
hist_212_agg = hist_212.groupby(pd.Grouper(freq='1D')).mean()
hist_212_agg.head()
```

```python
hist_212_agg['pm1_next_day'] = hist_212_agg['pm1'].shift(-1)
hist_212_agg['pm25_next_day'] = hist_212_agg['pm25'].shift(-1)
hist_212_agg['pm10_next_day'] = hist_212_agg['pm10'].shift(-1)

hist_212_agg.head()
```

```python
hist_212_agg = hist_212_agg.dropna()
hist_212_agg.head()
```

### Scaling

```python
from sklearn.preprocessing import StandardScaler
hist_212_agg.iloc[:] = StandardScaler().fit_transform(hist_212_agg)
hist_212_agg.head()
```

```python
X = hist_212_agg[['temperature', 'humidity', 'pressure', 'pm1', 'pm25', 'pm10']]
y = hist_212_agg[['pm1_next_day', 'pm25_next_day', 'pm10_next_day']]
```

## Train/test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train
```

## Regressions

```python
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures

class SameValueRegression:
    def predict(self, X):
        return X[['pm1', 'pm25', 'pm10']].to_numpy()
```

### Training

```python
pol2_feature_trans = PolynomialFeatures(2).fit(X_train)
pol2_features = pol2_feature_trans.transform(X_train)
pol2_reg = Ridge(alpha=1.0, solver='lsqr').fit(pol2_features, y_train)

pol3_feature_trans = PolynomialFeatures(3).fit(X_train)
pol3_features = pol3_feature_trans.transform(X_train)
pol3_reg = Ridge().fit(pol3_features, y_train)

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

mlp_reg = MLPRegressor(max_iter=1000)
mlp_reg.fit(X_train, y_train)

same_value_reg = SameValueRegression()
```

### Predictions

```python
X_test_pol2 = pol2_feature_trans.transform(X_test)
y_pred_pol2 = pol2_reg.predict(X_test_pol2)

X_test_pol3 = pol3_feature_trans.transform(X_test)
y_pred_pol3 = pol3_reg.predict(X_test_pol3)

y_pred_lin = lin_reg.predict(X_test)

y_pred_mlp = mlp_reg.predict(X_test)

y_pred_same_val = same_value_reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
# print('Mean squared error: %.2f' % (mean_squared_error(y_test, y_pred))

print('Coefficient of determination - polinomial 2: %.2f' % r2_score(y_test, y_pred_pol2))

print('Coefficient of determination - polinomial 3: %.2f' % r2_score(y_test, y_pred_pol3))

print('Coefficient of determination - linear: %.2f' % r2_score(y_test, y_pred_lin))

print('Coefficient of determination - mlp: %.2f' % r2_score(y_test, y_pred_mlp))

print('Coefficient of determination - same val: %.2f' % r2_score(y_test, y_pred_same_val))
```

```python
print('Mean squared error - polinomial 2: %.2f' % mean_squared_error(y_test, y_pred_pol2))

print('Mean squared error - polinomial 3: %.2f' % mean_squared_error(y_test, y_pred_pol3))

print('Mean squared error - linear: %.2f' % mean_squared_error(y_test, y_pred_lin))

print('Mean squared error - mlp: %.2f' % mean_squared_error(y_test, y_pred_mlp))

print('Mean squared error - same val: %.2f' % mean_squared_error(y_test, y_pred_same_val))
```
